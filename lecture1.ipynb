# Lecture 1: The Nature of Classical Physics

**Important Notions: What Is Classical Physics**

* Classical Physics are the fields in physics that were before quantum physics. These include:
  * Newtonian Mechanics
  * Maxwell-Faraday Electromagnetism
  * Einstein's General Theory of Relativity

* Physicist _Pierre-Simone Laplace_ claimed that if an "intellect" knew:
  * all the forces that set nature into motion, 
  * all the positions of all items present in nature, 
  * and was able to analyse and process all the relevant information
  
  this "intellect" would be able to make sense of the past and predict the future. (quote paraphrased)


* **Determinism:** a situation in classical mechanics where you can predict the future of a physical _system_ given that you:
  * have perfect information about the system in question...
  * ... and know the equations that govern the behaviour of said system.
* **Reversible:** we say that a system is reversable if we can use the same equations to explain the past

**Important Notions: Simple Dynamical Systems and the Space of States**

* **System:** a collection of objects - examples are: 
    * particles,
    * fields,
    * waves...
    * ...and similar entities.

* **Closed System:** a kind of system that encompasses the entire universe or is an entity thats behaviour is not affected by outside influences.

**Exercise 1:** Since the notion is so important to theoretical physics, think about what a closed system is and speculate on whether closed systems can actually exist. What assumptions are implicit in establishing a closed system? What is an open system?

**Answer\*:** A closed system is a physical entity where its are consistent, is self-contained and is not affected by outside forces. An open system on the other hand, is a physical system where its laws are consistent, but isn't self contained and is affected by outside forces.

### Interlude 1: Spaces, Trigonometry, and Vectors

**Exercise 1:** Using a graphing calculator or a program like _Mathematica_, plot each of the following functions. See the next section if you are unfamiliar with the trigonometric functions.

$f(t) = t^4 + 3t^3 - 12t^2 + t - 6$

$g(x) = sin \ x - cos \ x$

$\theta(\alpha) = e^\alpha + \alpha \ ln \ \alpha$

$x(t) = sin^2 \ t - cos \ t$

```python
# Our solution to Interlude 1's first exercise
#  We're gonna use Python's NumPy and matplotlib 
#  to graph these equations.

import numpy
import matplotlib.pyplot as plt

x_s = 0.01

def figure_plotter(f, x_i, x_f, x_s, title, xlabel, ylabel):
    plt.figure()
    plt.plot(numpy.arange(x_i, x_f + 1, x_s), [f(x) for x in numpy.arange(x_i, x_f + 1, x_s)])
    plt.title("Graph of: " + title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

f = lambda t: t**4 + 3 * t**3 - 12 * t**2 + t - 6
g = lambda x: numpy.sin(x) - numpy.cos(x)
theta = lambda alpha: numpy.exp(alpha) + alpha * numpy.log(alpha)
x = lambda t: (1./2 - 1./2 * numpy.cos(2 * t)) - numpy.cos(t)

figure_plotter(f, -4, 4, x_s, "$f(t)$", "$t$", r"$f(t) = t^4 + 3t^3 - 12t^2 + t - 6$")
figure_plotter(g, 0, 2 * 3.14, x_s, "$g(x)$", "$x$", "$g(x) = sin \ x - cos \ x$")
figure_plotter(theta, 0, 5, x_s, r'$\theta (a)$', r'$\alpha$', r'$\theta(\alpha) = e^\alpha + \ \alpha ln \ \alpha$')
figure_plotter(x, 0, 2 * 3.14, x_s, "$x(t)$", "$t$", r'$x(t) = sin^2 \ t - cos \ t$')

plt.show()
```
